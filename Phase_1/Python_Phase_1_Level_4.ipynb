###1. Take a character and check if it is a letter, a digit, or neither.

```python
x = input("Enter the character: ")

if x.isalpha():
  print("Letter")
elif x.isdigit():
  print('Digit')
else:
  print('Neither')
```

### 2. Take a number and print “Fizz” if divisible by 3, “Buzz” if divisible by 5, and “FizzBuzz” if divisible by both.

```python
x = int(input("Enter the number: "))

if x % 3 == 0 and x % 5 == 0:
  print("FizzBuzz")
elif x % 5 == 0:
  print("Buzz")
elif x % 3 == 0 :
  print("Fizz")
```

### 4. Take 24-hour time (hours and minutes) and print whether it is AM or PM.

```python
x,y = map(int, input("Enter the time in this format HH:MM (0-24): ").split(':'))

if x in range(0,12) and y in range(0,60):
  print("Its AM")
elif x in range(12,25) and y in range(0,60):
  print("its PM")
else:
  print("Please mention proper time.")

#One - liner

x, y = map(int, input("Enter the time in this format HH:MM (0-24): ").split(':'));
print("AM" if x in range(0,12) and y in range(0,60) else "PM" if x in range(12,25) and y in range(0,60) else "Invalid Time" )
```

### 8. Take a weekday number (1–7) and determine if it is a weekday or weekend.

```python
x = int(input("Enter the weekday number (1-7): "))
y = lambda x: "Weekday" if x in range(1,6) else "Weekend"
print(y(x))
```

###10. Take a password string and check basic rules (length ≥ 8 and contains at least one digit).

```python
x = str(input("Enter your password: "))

if len(x) >= 8 and any(i.isdigit() for i in x):
  print("Valid")
else:
  print("invalid")
```
