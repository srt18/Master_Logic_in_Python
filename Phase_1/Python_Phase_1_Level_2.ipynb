### 1. Take three sides and check if they form a valid triangle.

```python
a = int(input("Enter the side a: "))
b = int(input("Enter the side b: "))
c = int(input("Enter the side c: "))

if (a+b)>c and (b+c)>a and (c+a)>b:
  print("Valid Triangle")
else:
  print("Not a valid triangle")
```

###2. If the sides form a valid triangle, determine whether it is equilateral,isosceles, or scalene.

```python
""" """
a = int(input("Enter the side a: "))
b = int(input("Enter the side b: "))
c = int(input("Enter the side c: "))

if (a+b)>c and (b+c)>a and (c+a)>b:
  if a==b==c:
    print("Valid triangle and its equilateral triangle")
  elif a==b or b==c or a==c:
    print("Valid triangle and its isosceles triangle")
  else:
    print("Valid triangle and its scalene triangle")
else:
  print("Not a valid triangle")
```

###3. Take marks (0–100) and print the corresponding grade (A/B/C/D/F).

###7. Take two numbers and determine whether both are even, both are odd, or one is even and one is odd.

```python
x = int(input("Enter number x: "))
y = int(input("Enter number y: "))

if (x%2 == 0) and (y%2 == 0):
  print("Both are even")
elif (x%2 != 0) and (y%2 == 0):
  print("x is odd and y is even")
elif (x%2 == 0) and (y%2 != 0):
  print("y is odd and x is even")
else:
  print("Both are odd")
```

###8. Take an alphabet character and check if it lies between ‘a’ and ‘m’ or ‘n’ and ‘z’.

```python
ch = input("Enter the alphabet: ")

if ('a'<= ch <='m'):
  print("Alphabet lies between a and m")
elif ('n'<= ch <='z'):
  print("Alphabet lies between n and z")
else:
  print("Character not lies between a to z")

"""Explaination : Characters have ASCII values, and Python compares them lexicographically.
If the character is outside lowercase alphabets, it returns a message accordingly. """
```
