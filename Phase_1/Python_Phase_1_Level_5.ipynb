###1. Take coordinates (x, y) and check if the point lies on the X-axis, Y-axis, or at the origin.

```python
x, y = map(int, input("Enter the coordinates (x,y): ").split(","))

if x == 0  and y == 0 :
  print("Point at origin")
elif y == 0:
  print("Lies on the X-axis")
elif x == 0:
  print("Lies on the Y-axis")
else:
  print("Somewhere else")
```

###2. Take three numbers and check if they can form a Pythagorean triplet.

```python
x,y,z = map(int, input("Enter the three numbers for Pythagorean triplet: ").split(","))

if (x**2 + y**2 == z**2) or (y**2 + z**2 == x**2) or (z**2 + x**2 == y**2):
  print("They are Pythagorean triplet")
else:
  print("There are not a triplets")
```

###3. Take day and month and check if it forms a valid calendar date (ignoring leap years).

```python
x, y = map(int, input("Enter the days and month: ").split(","))

days_in_months = {
    1: 31, 2: 28, 3: 31, 4: 30,
    5: 31, 6: 30, 7: 31, 8: 31,
    9: 30, 10: 31, 11: 30, 12: 31
}

if y < 1 or y > 12:
  print("Please enter valid days and months")
elif 1 <= x <= days_in_months[y]:
  print("Valid")
else:
  print("Invalid")
```

###4. Take time (hours and minutes) and print the smaller angle between the hour and minute hands.


```python
x, y = map(int, input("Enter the hours and minutes: ").split(":"))

if 1 <= x <= 12 and 1 <= y <= 60:
  z = y*6
  print(f"Smaller angle between hour and minute hands is:",z)
else:
  print("Please enter valid time.")
```

###5. Take three numbers and check if they are in arithmetic progression.

```python
x, y, z = map(int, input("Enter three numbers: ").split(","))

if (y-x) == (z-y) or (x-y) == (z-x) or (z-x) == (y-z):
  print("They are arithmetic progression")
else:
  print("They are not in arithmetic progression")
```

###6. Take three numbers and check if they are in geometric progression.

```python
x, y, z = map(int, input("Enter three numbers: ").split(","))

if y*y == x*z:
  print("They are geometric progression")
else:
  print("They are not geometric progression")
```

###7. Take a 3-digit number and check if the sum of the first and last digit equals the middle digit.


```python
x, y, z = map(int, input("Enter three numbers: ").split(","))

if (x+z) == y:
  print("Yes")
else:
  print("No")
```

###8. Take an integer (1–9999) and check if the sum of its digits is greater than the product of its digits.


```python
x = list(input("Enter the number: "))
digits = [int(d) for d in x ]

if sum(digits) > eval("*".join(x)):
  print("Yes, sum of its digits is greater than the product of its digits.")
elif sum(digits) == eval("*".join(x)):
  print("Sum is equal to product")
else:
  print("Sum is not greater than the product of its digits")
```

###9. Take two dates (day and month) and determine which one comes first in the calendar.


```python
date_1, month_1 = map(int, input("Enter the first date: ").split("/"))
date_2, month_2 = map(int, input("Enter the second date: ").split("/"))

if date_2 == date_1 and month_1 == month_2:
  print("Both are same date")
elif month_2 > month_1:
  print(f"This date comes first in the calendar: ",date_1,"/",month_1)
elif date_1 > date_2:
  print(f"This date comes first in the calendar: ",date_2,"/",month_2)
elif month_1 > month_2:
  print(f"This date comes first in the calendar: ",date_1,"/",month_1)
else:
  print(f"This date comes first in the calendar: ",date_2,"/",month_2)
```

###10. Take a year and print the corresponding century (e.g., “19th century”, “20th century”)

```python
x = int(input("Enter the year:"))

if x in range(1900,2000):
  print("20th century")
elif x in range(1800,1900):
  print("19th century")
else:
  print("Before 19th century")
```
