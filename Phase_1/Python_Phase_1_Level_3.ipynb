### 1. Take a 3-digit number and check if all digits are distinct.

```python
# if else
x = int(input("Enter the 3-digit number: "))

a = x// 100
b = (x //10) % 10
c = x % 10
if a != b and b != c and a != c:
  print("All are digits are distinct")
else:
  print("Not distinct")

# One liner
print("Distinct" if len(set(str(int(input("Enter the 3-digit number: "))))) == 3 else "Not distinct")
```

### 2. Take a 3-digit number and determine if the middle digit is the largest, smallest, or neither.

```python
x = int(input("Enter the 3-digit number: "))

a = x//100
b = (x//10)%10
c = x % 10
if b < a and b < c:
  print("Its smallest")
elif b > a and b > c:
  print("Its largest")
else:
  print("Neither")
```

### 4. Check whether a given integer is single-digit, double-digit, or multi-digit.

```python
x = int(input("Enter the number: "))

if (x // 10) == 0:
  print("Single-digit")
elif (x//100) == 0:
  print("Double-digit")
else:
  print("Multi-digit")

```

###6. Take coordinates (x, y) and determine which quadrant the point lies in.

```python
x = int(input("Enter the x coordinates: "))
y = int(input("Enter the y coordinates: "))

if x > 0 and y > 0:
  print("1st quadrant")
elif x < 0 and y > 0:
  print("2nd quadrant")
elif x < 0 and y < 0:
  print("3rd quadrant")
elif x > 0 and y < 0:
  print("4th quadrant")
elif x == 0 and y == 0:
  print("Origin")
else:
  print("Lies either on x or y axis")
```

###7. Check if an amount can be evenly divided into 2000, 500, and 100 currency notes.


```python
x = int(input("Enter the amount:"))

if x%2000 == 0:
  print("evenly divided 2000")
elif x%500 == 0:
  print("evenly divided 500")
elif x%100 == 0:
  print("Evenly divided 100")
else:
  print("Not divide by any currency")
```

###9. Take two angles of a triangle and compute the third angle.


```python
x = int(input("Enter the angle x:"))
y = int(input("Enter the angle y:"))

if (x+y) < 180 and x != 0 and y != 0:
  z = 180 - (x+y)
  print("Third angle",z)
else:
  print("Its not a triangle")
```
