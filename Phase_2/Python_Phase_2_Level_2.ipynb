###1. Count the number of digits in a given number.


```python
x = int(input("Enter the number: "))
count = 0

if x == 0:
  count = 1
else:
  while x > 0 :
    x//=10 # Division by 10 in integer
    count +=1

print("Count of the given number: ",count)
```

###2. Print the reverse of a given number.


```python
num = int(input("Enter the number: "))
reverse = 0
while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num //=10

print(reverse)

#One-liner
x = str(input("Enter the number: "))
print(x[::-1])
```

###3. Check if a number is a palindrome.


```python
num = int(input("Enter the number: "))
original = num
reverse = 0
while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num //=10
if original == reverse:
  print("Yes its palindrome")
else:
  print("No its not")
```

###4. Find the sum of digits of a number.

```python
num = int(input("Enter a number: "))
total = 0

while num > 0:
    digit = num % 10
    total += digit
    num //= 10

print("Sum of digits:", total)
```

###5. Check if a number is an Armstrong number.


```python
num = int(input("Enter a number: "))
original = num
total = 0
temp = num
while temp > 0:
    temp //= 10
    total += 1

sum_of_power = 0
temp_power_sum = num
while temp_power_sum > 0:
    digit = temp_power_sum % 10
    sum_of_power += digit **  total
    temp_power_sum//=10

if sum_of_power == original:
  print("Its a an armstrong number")
else:
  print("Its not armstrong number")
```

###6. Check if a number is a perfect number.


```python
x = int(input("enter the number: "))
total = 0
for i in range(1,x):
  if x % i == 0:
    total += i

if total == x:
  print("Its prefect number")
else:
  print("Its not prefect number")
```

###7. Print all prime numbers between 1 and 100.


```python
for i in range(2, 101):
    prime = True
    for d in range(2, int(i**0.5) + 1):
      if i%d == 0:
        prime = False
        break
    if prime:
      print(i,end=" ")
```

###8. Check if a number is prime or not.


```python
x = int(input("Enter the number: "))

if x < 2:
    print("Its not prime")
else:
    for d in range(2, int(x**0.5)+1):
      if x%d == 0:
        print("its not prime")
        break
    else:
        print("its prime")
```
