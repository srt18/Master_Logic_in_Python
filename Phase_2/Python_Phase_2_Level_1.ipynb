###1. Print numbers from 1 to 10.

```python
for x in range(1,11):
  print(x,end=" ")
```

###2. Print all even numbers between 1 and 50.

```python
for x in range(1,51):
  if x%2 == 0:
    print(x, end=" ")

print(  )
#one-liner list comprehensions
y = [x for x in range(1,51) if x%2 == 0 ]
print(y)
```

###3. Print all odd numbers between 1 and 50.

```python
for x in range(1,51):
  if x%2 != 0:
    print(x, end=" ")

print( )

#One-liner
print([x for x in range(1,51) if x%2 != 0], end=" ")
```

###4. Print numbers from 10 down to 1.

```python
for x in range(10,0,-1):
  print(x, end= " ")

print( )

#One-liner
print([x for x in range(10,0,-1)], end=" ")
```

###7. Print the sum of all even numbers up to n.

```python
n = int(input("enter the even number sum you want: "))
total = 0
for i in range(2, n+1, 2):
  total += i
  continue
print("Sum of all even numbers are", total)

#One-liner
print( )
n = int(input("enter the even number sum you want: "))
print("Suum of all even numbers is:", sum(range(2, n+1, 2)))
```

###8. Print the sum of all odd numbers up to n.

```python
n = int(input("enter the odd number sum you want: "))
total = 0
for i in range(1, n+1, 2):
  total += i
  continue
print("Sum of all odd numbers are", total)

#One-liner
print( )
n = int(input("enter the odd number sum you want: "))
print("Suum of all odd numbers is:", sum(range(1, n+1, 2)))
```

###9. Print the factorial of a given number.

```python
x = int(input("enter the number: "))
fact = 1
for i in range(1,x):
  fact *= i
print("Factorial of given number:", fact)

# Using Recursion
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

n = int(input("Enter a number: "))
print("Factorial of", n, "is:", factorial(n))
```
