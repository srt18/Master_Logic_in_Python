###1. Print the squares of numbers from 1 to n.


```python
n = int(input("enter the n numbers:"))

for i in range(1,n+1):
  sq = i ** 2
  print(sq, end=" ")

#one-liner version
print( )
print(*((lambda x: x**2)(i) for i in range(1,(int(input("enter the n numbers:")))+1)), end=" ")
```

###2. Print cubes of numbers from 1 to n.

###3. Print all numbers between a and b divisible by 7.


```python
a,b = map(int, input("Enter the numbers between a and b: ").split(","))

for i in range(a,b+1):
  if i % 7 == 0:
    print(i, end=" ")
```

###4. Find HCF (GCD) of two numbers using loops.

```python
# Euclidean Algorithm for GCD
a,b = map(int, input("Enter the two numbers: ").split(","))

while b!=0:
  a, b = b, a%b

print("HCF: ",a)
```

###5. Find LCM of two numbers using loops.


```python
a,b = map(int, input("Enter the two numbers: ").split(","))

x, y = a, b
while b!=0:
  a, b = b, a%b

hcf = a
lcm = (x*y)/hcf

print("LCM: ",lcm)
```

###6. Print all factors of a given number.


```python
n = int(input("enter the number:"))

for i in range(1, n+1):
  if n % i == 0:
    print(i, end=" ")
```

###7. Find the sum of all factors of a number.


```python
n = int(input("enter the number:"))
total = 0
for i in range(1, n+1):
  if n % i == 0:
    total += i

print("Sum of all factors of a number: ",total, end=" ")
```

###8. Check if a number is a strong number (sum of factorials of digits = number).


```python
x = (input("enter the number: "))

def fact(n):
  fact = 1
  for i in range(1, n+1):
    fact *= i
  return fact

temp = int(x)
sum_fact = 0

while temp > 0:
    digits = temp%10
    sum_fact += fact(digits)
    temp//=10
    # print(sum_fact)

if sum_fact == int(x):
  print("Strong number is ", x)
else:
  print("not")
```

###9. Print first n terms of an arithmetic progression (a, d).

```python
a, d = map(int, input("Enter the arithmetic progression a, d: ").split(","))
n = int(input("Enter the first n terms: "))
for i in range(0,n):
  t = a + i*d
  print(t,end=" ")
```

###10. Print first n terms of a geometric progression (a, r).

```python
a, r = map(int, input("Enter the geometric progression a, r: ").split(","))
n = int(input("Enter the first n terms: "))
for i in range(0,n):
  t = a*(r**i)
  print(t,end=" ")
```
